Tipo de Registro

```python
# TODO: Validación tipo registro
dfc['Tipo de registro_'] = [False if i == 2 else True for i in dfc['Tipo de registro']]
```

Tipo de indentificacion del usuario

```python
# TODO: Obtener indentificaciones validas
tip_id_val = ['SC', 'SV', 'PE', 'PT', 'AS', 'CC', 'CN', 'NV', 'MS', 'TI', 'RC', 'CE', 'PA']

is_val = []
for ti in dfc['Tipodeidentificacióndelusuario']:
    if ti in tip_id_val:
        is_val.append(False)
    else:
        is_val.append(True)

dfc['Tipodeidentificacióndelusuario_'] = is_val
```

```python
# TODO: Validación de Numero de Indentificación
dfc['Númerodeidentificacióndelusuario_'] = dfc['Númerodeidentificacióndelusuario'].duplicated(keep=False)
```
